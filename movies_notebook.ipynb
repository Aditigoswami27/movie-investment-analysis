```python
movies= movies.rename(columns={
    "movie_title": "title",
    "title_year": "year",
    "gross": "revenue",
    "num_voted_users": "votes_count",
    "imdb_score": "imdb_rating"
})
```

```python
movies["country"] = movies["country"].fillna("Unknown")
movies["language"] = movies["language"].fillna("Unknown")
```

```python
movies = movies.dropna()
```

```python
movies= movies.rename(columns={
    "title_year": "year"
})
```

```python
movies["year"] = movies["year"].astype(int)
movies["duration"] = movies["duration"].astype(int)
```

```python
movies["profit"] = movies["revenue"] - movies["budget"]
movies["roi"] = movies["profit"] /movies["budget"]
movies
```

```python
movies["roi"] =movies["roi"].round(2)
```

```python
movies[movies["budget"] > 1000000000]
```

```python
movies= movies[movies["budget"] < 1000000000]
```

```python
movies.budget.describe()
```

```python
def budget_category(cell):
    if cell < 10000000:
        return "Low budget"
    if cell >= 10000000 and cell<=50000000:
        return "Mid budget"    
    if cell > 50000000:
        return "High budget"  
movies["budget_category"] =movies["budget"].apply(budget_category)        
```
